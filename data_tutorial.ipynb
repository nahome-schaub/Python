```python
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
my_data = pd.read_csv("gdp.csv")
my_data
```

```python
def get_country_data(country, data = my_data):
    country_data = my_data[my_data["LOCATION"] == country]
    return country_data

def get_avg_gdp(country, data = my_data):
    values = [value for value in country["Value"]]
    return sum(values) / len(values)
```

```python
my_countries = ["USA", "AUS", "FIN"]
for country in my_countries:
    print(country)
    current_data = get_country_data(country)
    avg_gdp = get_avg_gdp(current_data)
    gdp_array = [value for value in current_data["Value"]]
    print(f"GDP 2022: {gdp_array[-1]}")
    print(f"GDP 1960: {gdp_array[0]}")
    print(f"Average GDP : {avg_gdp}\n")
```
